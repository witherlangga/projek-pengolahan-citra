```python
## Segmentasi & Klasifikasi Warna Daun dengan Visualisasi per Cluster

from rembg import remove
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os, io
from sklearn.cluster import KMeans
import colorsys

## Folder Input dan Output
input_folder = "img"
output_folder = "hasil_segmentasi"
os.makedirs(output_folder, exist_ok=True)

## Fungsi Klasifikasi Warna Daun
def classify_leaf_color(rgb_color):
    """
    Mengklasifikasikan warna daun berdasarkan nilai HSV.
    Input: rgb_color = array [R, G, B]
    Output: kategori kondisi daun
    """
    r, g, b = rgb_color / 255.0
    h, s, v = colorsys.rgb_to_hsv(r, g, b)
    h_deg = h * 360

    # Daun hijau sehat (gelap dan jenuh)
    if 90 <= h_deg <= 150 and s >= 0.25 and v >= 0.25:
        return "Daun Sehat (Hijau Tua)"

    # Kuning kehijauan - awal penyakit
    elif 60 <= h_deg < 90 and s >= 0.25 and v >= 0.3:
        return "Daun Sakit Ringan (Kuning Kehijauan)"

    # Kuning atau coklat muda - penyakit sedang
    elif 30 <= h_deg < 60 and s >= 0.15 and v >= 0.25:
        return "Daun Sakit Sedang (Kuning / Coklat Muda)"

    # Coklat tua / hitam - penyakit berat atau mati
    elif (v < 0.35 and s < 0.5) or (h_deg < 30 and v < 0.4):
        return "Daun Rusak Berat (Coklat / Hitam)"

    # Warna lain (abu-abu, putih, noise)
    else:
        return "Lainnya"

## Loop Semua File Gambar di Folder Input
for filename in os.listdir(input_folder):
    if not filename.lower().endswith((".jpg", ".jpeg", ".png")):
        continue

    print(f"\nMemproses gambar: {filename}")
    image_path = os.path.join(input_folder, filename)

    # 1. Hapus background
    with open(image_path, "rb") as f:
        input_image = f.read()
    output_image = remove(input_image)
    result = Image.open(io.BytesIO(output_image)).convert("RGBA")

    # 2. Konversi ke Numpy
    image_rgba = np.array(result)
    rgb = image_rgba[:, :, :3]
    alpha = image_rgba[:, :, 3]
    mask_leaf = alpha > 128
    pixels_leaf = np.float32(rgb[mask_leaf])

    if pixels_leaf.shape[0] == 0:
        print("Tidak ada piksel daun terdeteksi, lewati gambar ini.")
        continue

    # 3. K-Means Clustering
    K = 5
    kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
    labels = kmeans.fit_predict(pixels_leaf)
    centers = np.uint8(kmeans.cluster_centers_)

    # 4. Klasifikasi warna daun tiap cluster
    labels_keterangan = [classify_leaf_color(c) for c in centers]

    # 5. Rekonstruksi hasil segmentasi
    segmented_image = np.zeros_like(rgb)
    segmented_image[mask_leaf] = centers[labels]

    height, width, _ = rgb.shape
    label_image = np.zeros((height, width), dtype=np.uint8)
    label_image[mask_leaf] = labels

    # 6. Visualisasi hasil utama
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(rgb)
    plt.title("Daun Tanpa Background")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(segmented_image)
    plt.title(f"Hasil Segmentasi (K={K})")
    plt.axis("off")
    plt.show()

    # 7. Simpan hasil segmentasi utama
    base_name = os.path.splitext(filename)[0]
    segmented_path = os.path.join(output_folder, f"{base_name}_segmentasi.png")
    Image.fromarray(np.uint8(segmented_image)).save(segmented_path)


    # 8. Visualisasi Setiap Cluster Secara Terpisah

    cluster_viz_folder = os.path.join(output_folder, f"{base_name}_cluster")
    os.makedirs(cluster_viz_folder, exist_ok=True)

    for i, (color, desc) in enumerate(zip(centers, labels_keterangan)):
        cluster_mask = (label_image == i)
        cluster_color = np.zeros_like(rgb)
        cluster_color[cluster_mask] = rgb[cluster_mask]

        plt.figure(figsize=(5, 5))
        plt.imshow(cluster_color)
        plt.title(f"Cluster {i+1}  {desc}")
        plt.axis("off")

        # Bersihkan nama file agar aman
        safe_desc = (
            desc.replace(' ', '_')
                .replace('/', '-')
                .replace('(', '')
                .replace(')', '')
        )

        cluster_file = os.path.join(
            cluster_viz_folder,
            f"{base_name}_cluster{i+1}_{safe_desc}.png"
        )

        plt.savefig(cluster_file, bbox_inches="tight", pad_inches=0.1)
        plt.close()

        print(f"Cluster {i+1} disimpan: {os.path.basename(cluster_file)} ({desc})")

    print(f"Selesai memproses: {filename}")
    print(f"Hasil disimpan di folder: {output_folder}")

print("\nSemua gambar selesai diproses.")
```
