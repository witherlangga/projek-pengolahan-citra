```python
# =========================================================
# ðŸŒ¿ Batch Processing â€” Segmentasi & Klasifikasi Daun (Folder)
# Versi sama dengan "PC_Kelompok_9.ipynb", tapi otomatis untuk semua gambar
# =========================================================

# !pip install rembg opencv-python-headless numpy pillow scikit-learn matplotlib tqdm pandas --quiet

import os
import glob
import numpy as np
import pandas as pd
import cv2
from tqdm import tqdm
from rembg import remove
from PIL import Image
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import io
```

```python

# ===================== 1ï¸âƒ£ SETUP FOLDER =====================
INPUT_DIR = "img/"      # ðŸ‘‰ ganti dengan folder gambar kamu
OUTPUT_DIR = "hasil/"     # hasil akan disimpan di sini
os.makedirs(OUTPUT_DIR, exist_ok=True)
os.makedirs(os.path.join(OUTPUT_DIR, "visualisasi"), exist_ok=True)
```

```python

# ===================== 2ï¸âƒ£ FUNGSI KLASIFIKASI =====================
def classify_leaf_color(rgb_color):
    r, g, b = rgb_color
    if g > r and g > b:
        return "Daun Sehat (Hijau)"
    elif r > 150 and g > 150 and b < 100:
        return "Daun Sedikit Sakit (Kuning)"
    elif r > g and r > b and r > 120:
        return "Daun Sakit (Coklat Kekuningan)"
    elif r < 90 and g < 90 and b < 90:
        return "Daun Rusak Parah (Gelap)"
    else:
        return "Lainnya"
```

```python

# ===================== 3ï¸âƒ£ PROSES PER GAMBAR =====================
records = []

for img_path in tqdm(glob.glob(os.path.join(INPUT_DIR, "*.*"))):
    try:
        name = os.path.splitext(os.path.basename(img_path))[0]
        # --- Hapus background ---
        with open(img_path, "rb") as inp:
            result = remove(inp.read())
        image_rgba = np.array(Image.open(io.BytesIO(result)).convert("RGBA"))

        rgb = image_rgba[:, :, :3]
        alpha = image_rgba[:, :, 3]
        mask_leaf = alpha > 128
        pixels_leaf = rgb[mask_leaf].astype(np.float32)

        if pixels_leaf.shape[0] < 10:
            print(f"{name}: tidak ada piksel daun terdeteksi, dilewati.")
            continue

        # --- K-Means (K = 4) ---
        K = 4
        kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
        labels = kmeans.fit_predict(pixels_leaf)
        centers = np.uint8(kmeans.cluster_centers_)

        # --- Klasifikasi warna tiap cluster ---
        labels_keterangan = [classify_leaf_color(c) for c in centers]

        # --- Buat segmented image ---
        segmented_image = np.zeros_like(rgb)
        segmented_image[mask_leaf] = centers[labels]

        # --- Label per piksel untuk visualisasi cluster ---
        height, width, _ = rgb.shape
        label_image = np.zeros((height, width), dtype=np.uint8)
        label_image[mask_leaf] = labels

        # ===================== 4ï¸âƒ£ VISUALISASI =====================
        fig = plt.figure(figsize=(10, 12))
        plt.subplot(2, 2, 1)
        plt.imshow(rgb)
        plt.title("Gambar Daun Asli (Tanpa Background)")
        plt.axis("off")

        plt.subplot(2, 2, 2)
        plt.imshow(segmented_image)
        plt.title(f"Hasil Segmentasi Daun (K = {K})")
        plt.axis("off")

        # tampilkan tiap cluster
        for i, (color, desc) in enumerate(zip(centers, labels_keterangan)):
            cluster_mask = np.zeros((height, width, 3), dtype=np.uint8)
            cluster_mask[label_image == i] = color
            plt.subplot(2, 2, 3 + (i % 2))
            plt.imshow(cluster_mask)
            plt.title(f"Cluster {i+1} â€” {desc}")
            plt.axis("off")

        plt.tight_layout()
        save_path = os.path.join(OUTPUT_DIR, "visualisasi", f"{name}_hasil.png")
        plt.savefig(save_path, bbox_inches="tight")
        plt.close(fig)

        # ===================== 5ï¸âƒ£ SIMPAN KE CSV =====================
        for i, (color, desc) in enumerate(zip(centers, labels_keterangan)):
            records.append({
                "Nama File": name,
                "Cluster": i + 1,
                "RGB": str(tuple(color)),
                "Klasifikasi": desc
            })

    except Exception as e:
        print(f"âŒ Error di {img_path}: {e}")

# ===================== 6ï¸âƒ£ RINGKASAN =====================
df = pd.DataFrame(records)
csv_path = os.path.join(OUTPUT_DIR, "hasil_klasifikasi_daun.csv")
df.to_csv(csv_path, index=False)
print(f"\nâœ… Selesai! Semua hasil disimpan di: {OUTPUT_DIR}")
print(f"ðŸ“„ File klasifikasi: {csv_path}")
print(f"ðŸ–¼ï¸ Visualisasi per gambar: {os.path.join(OUTPUT_DIR, 'visualisasi')}")
```
